```python
import numpy as np
import matplotlib.pyplot as plt

data = np.arange(0,1.1,0.01)
plt.title("lines")
plt.xlabel("x")
plt.ylabel("y")
plt.xlim((0,1))
plt.ylim((0,1))
plt.xticks(np.arange(0,1,0.2))
plt.yticks(np.arange(0,1,0.2))
plt.plot(data, data**2)
plt.plot(data, data**4)
plt.legend(['y=x^2', 'y=x^4'])
plt.show()
```

## 动态rc参数

- pyplot使用rc配置文件来自定义图形的各种默认属性
- 窗口大小、每英寸点数、线宽、颜色和样式、坐标轴、网格属性、文本、字体等
- 可以动态修改rc参数
- 存储在字典的rc参数称为rcParams
- 修改后，绘图的默认参数会改变

```python
## 原图
x = np.linspace(0, 4*np.pi)## 生成x轴数据
y = np.sin(x)## 生成y轴数据
plt.plot(x,y,label="$sin(x)$")## 绘制sin曲线图
plt.title('sin')
plt.savefig('./matplotlib/默认sin曲线.png')
plt.show()
```

```python
## 修改rc参数后的图
plt.rcParams['lines.linestyle'] = '-.'
plt.rcParams['lines.linewidth'] = 3
plt.plot(x,y,label="$sin(x)$")## 绘制三角函数
plt.title('sin')
plt.savefig('./matplotlib/修改rc参数后sin曲线.png')
plt.show()
```

rc参数名称|解释|取值
:-:|:-:|:-:|
lines.linestyle|样式|`- -- -. :`
lines.linewidth|宽度|0~10
lines.marker|点的形状|'o D h . , S '等
lines.markersize|点的大小|0~10


- pyplot字体不支持中文，要修改字体
- 修改字体后，坐标轴部分字符无法显示，要修改axes.unicode_minus

```python
## 无法显示中文标题
plt.plot(x,y,label="$sin(x)$")## 绘制三角函数
plt.title('sin曲线')
plt.savefig('./matplotlib/无法显示中文标题sin曲线.png')
plt.show()
```

```python
##设置rc参数显示中文标题
## 设置字体为SimHei显示中文
plt.rcParams['font.sans-serif'] = 'SimHei'
plt.rcParams['axes.unicode_minus'] = False ## 设置正常显示符号
plt.plot(x,y,label="$sin(x)$")## 绘制三角函数
plt.title('sin曲线')
plt.savefig('./matplotlib/显示中文标题sin曲线.png')
plt.show()
```
